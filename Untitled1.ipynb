```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import os

# Data for the timeline
data = {
    'Task': ['Market Research ', 'Develop Menu and Pricing', 'Establish patnership with local suppliers','Optimise Price','Marketing Campaign','launch Tacos in selected stores','Expansion'],
    'Start': pd.to_datetime(['2024-07-01', '2024-08-01', '2024-09-01', '2024-10-01', '2024-11-01', '2024-12-01', '2025-02-01']),
    'End': pd.to_datetime(['2024-08-01', '2024-09-01', '2024-10-01', '2024-11-01', '2024-12-01', '2025-01-01', '2025-03-01'])
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot each task
for i, task in df.iterrows():
    ax.plot([task['Start'], task['End']], [i, i], linewidth=10, marker='o')

# Set y-ticks and labels
ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['Task'])

# Format x-axis dates
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Rotate date labels
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# Add gridlines and title
ax.grid(True)

plt.tight_layout()

# Path to the downloads folder with the filename
downloads_path = os.path.join("C:/Users/Muhammad Sarim/Downloads", "Timeline_for_Enhancing_Local_Preferences.png")

# Save the plot to the downloads folder
plt.savefig(downloads_path)

plt.show()
```
